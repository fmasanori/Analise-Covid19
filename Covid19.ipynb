```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import geopandas as gpd
plt.style.use('classic')
%matplotlib inline
```

```python
filename = r'HIST_PAINEL_COVIDBR_01jul2020.xlsx'
df = pd.read_excel(filename)
```

```python
df['data'] = df['data'].dt.strftime('%d/%m')
```

```python
df.query('municipio  == "São José dos Campos"').plot(kind='bar', x='data', y ='casosAcumulado',figsize = (20,10),label='Casos confirmados')
plt.title('Número de casos confirmados acumulados em São José dos Campos', fontsize=17)
plt.xlabel('Data')
plt.show()
```

```python
df.query('municipio  == "São José dos Campos"').groupby('semanaEpi')['casosNovos'].sum().plot(kind='bar', figsize = (20,10),label='Casos confirmados', color = 'm')
plt.title('Número de casos semanais em São José dos Campos',fontsize=17)
plt.xlabel('Semana')
plt.show()
```

```python
df.query('municipio == "São José dos Campos" and semanaEpi >= 23').plot(kind= 'bar', x = 'data', y= 'casosNovos',label = 'Número de casos confimados', figsize = (20,10))
plt.title('Novos casos nas últimas 5 semanas em São José dos Campos', fontsize=17)
plt.ylabel('Número de novos casos diários')
plt.xlabel('Semana')
plt.show()
```

```python
df.query('municipio == "São José dos Campos" and semanaEpi >= 23').plot(kind='bar', x ='data', y='obitosNovos',label = 'Número de óbitos',ax = plt.gca(), figsize = (20,10), color = 'RED')
plt.title('Novos óbitos nas últimas 5 semanas em São José dos Campos', fontsize=17)
plt.ylabel('Número de novos óbitos diários')
plt.xlabel('Semana')
plt.show()
```

```python
cidades = ['São José dos Campos', 'Jacareí', 'Caçapava', 'Taubaté']
cidades.sort()
for c in cidades:
    df[df['municipio'] == c].plot(kind='line', x = 'data', y='casosAcumulado', label = c, ax = plt.gca(), figsize = (20,10))
plt.title('Casos confirmados acumulados', fontsize=17)
plt.ylabel('Número de casos acumulados')
plt.xlabel('Data')
plt.show()
```

```python
dfsp = gpd.read_file('covid19-sp.csv', encoding='UTF-8')
dfsp = dfsp.drop("geometry", axis = 1)
dfsp = dfsp.drop([627], axis = 0)
dfsp["last_available_confirmed"] = dfsp["last_available_confirmed"].astype("int")
```

```python
dfsp = dfsp.merge(sp[["CD_MUN", "geometry"]], left_on="city_ibge_code", right_on="CD_MUN", how="left")
```

```python
dfsp.head()
```

```python
ax = sp.plot(color="grey", figsize = (20,10))
dfsp.plot(ax = ax, column='last_available_confirmed',cmap = 'tab20b')
fig, ax = plt.subplots(figsize=(20, 0.5))
fig.subplots_adjust(bottom=0.5)

norm = mpl.colors.Normalize(vmin=0, vmax=129328)
cmap = mpl.cm.tab20b
cb1 = mpl.colorbar.ColorbarBase(ax, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')

cb1.set_label('Casos confirmados no estado de SP',fontsize=17)
plt.show()
```
